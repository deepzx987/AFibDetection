# Import and define params

```python
import pandas as pd
import numpy as np
from scipy.io import loadmat
from func import *
from tqdm import tqdm_notebook as tqdm
df = pd.read_csv('REFERENCE-v3.csv',header=None)

dataframe = df.values
dataDir = 'training2017/'
output_dir = 'Final_Database/'
T=7
fs = 300
```

# AFIB Data Preparation

```python
afib = np.array(())
for i in range(len(dataframe)):
    if dataframe[i][1] == 'A':
        filename = dataframe[i][0]
        # print(filename)
        mat_data = loadmat(dataDir+filename+".mat") 
        data = mat_data['val'].squeeze()
        data = np.nan_to_num(data) # removing NaNs and Infs
        data = data - np.mean(data)
        data = data/np.std(data)
        data = butter_bandpass_filter(data, lowcut=5.0, highcut=45.0, fs=fs, order=1)
        # print(data)
        afib = np.append(afib,data)

overlap = 0.75
window = fs*T
stride = int((1-overlap)*window)
name = str(T)+'_Afib.csv'

final_db = np.zeros(window)
for i in range(0,afib.shape[0]-window,stride):
    print (i,'/',afib.shape[0])
    segmented_signal = afib[i:i+window]
    #print (segmented_signal.shape)
    segmented_signal = segmented_signal.reshape(1,segmented_signal.shape[0])
    #print (segmented_signal.shape)
    final_db = np.vstack((final_db,segmented_signal))
    #print (final_db.shape)
    
final_db = final_db[1:,:]
filename = output_dir + name
np.savetxt(filename,final_db,delimiter=',')
```
